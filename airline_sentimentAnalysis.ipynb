```python
rev_df['Airline'] = rev_df['Reviews'].apply(lambda x: 'Air Canada')
rev_df['Lowercase'] = rev_df['Reviews'].apply(lambda x: ' '.join(word.lower() for word in x.split()))
rev_df['PunctuationsRemoved'] = rev_df['Reviews'].str.replace('[^|\w\s]','')
rev_df['OtherTextsRemoved'] = rev_df['PunctuationsRemoved'].str.replace('trip verified','')
rev_df['OtherTextsRemoved'] = rev_df['OtherTextsRemoved'].str.replace('not verified','')
rev_df['StopWordsRemoved'] = rev_df['OtherTextsRemoved'].apply(lambda x: ' '.join(x for x in x.split() if x not in stop))

rev_df
```

## **Code for finding the frequency of negative words.**

```python
test_df =pd.read_excel('result.xlsx')

result = {
    '-1': 0,
    '0': 0,
    '1': 0
}
dic = test_df.to_dict(orient='records')

for row in dic:
  if 'fantastic' in row['Reviews'].lower():
    if float(row['Nikhil']) == -1.0:
      result['-1'] += 1
    elif float(row['Nikhil']) == 1.0:
      result['1'] += 1
    else:
       result['0'] += 1

print(result)
```

# **Since Spacy Algorithm was better at analyzing the sentiment of the reviews, this algorithm is chosen for the users to input a statement and see the sentiment analysis first hand**

```python
user_input = input("The text to check: ")
user_input.replace('[^|\w\s]','')
text_tokens = word_tokenize(user_input)
tokens_without_sw= [word for word in text_tokens if not word in stop]

def listToString(s): 
    
    # initialize an empty string
    str1 = " " 
    
    # return string  
    return (str1.join(s))

new_word = listToString(tokens_without_sw)
new = nlp(new_word)._.blob.polarity
if new == 0:
  print("The statement is neutral")
elif new >= 0.1:
  print("The statement is positive")
else: print("The statement is negative")
```
